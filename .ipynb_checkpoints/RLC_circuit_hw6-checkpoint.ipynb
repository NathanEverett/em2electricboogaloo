```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
def true(t,w,V=1,R=1,C=1,L=1):
    return V*w*R*L/np.sqrt((R-w**2*C*R*L)**2+(w*L)**2)*np.cos(w*t + np.arctan((R-(w*C)**2)/(w*L)))
```

```python
w = np.linspace(.001,30,10000)
plt.plot(w,true(t=0, w=w))
plt.title('Parallel RLC')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
```

```python
t = np.linspace(.001,30,10000)
w=np.pi/4
plt.plot(t,true(t=t, w=w))
plt.title('Parallel RLC')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.plot(t,np.cos(w*t))
```
