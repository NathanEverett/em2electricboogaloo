```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
def emf(t,a,B=1e-2,w=1e-2,v=5,x0=0):
    return w*(2*a*v*t+v*B+a*x0)
```

```python
plots = 4
t = np.linspace(0,1,plots)
a = 1e-4

for i in range(0,plots):
    plt.plot(t, emf(t,a), label = "a = {}".format(a) )
    a*=10
    
plt.title("Emf vs Time: Alpha")
plt.xlabel("Time")
plt.ylabel("EMF")
plt.legend()
```

```python
plots = 4
t = np.linspace(0,1,plots)
v = 1e-2

for i in range(0,plots):
    plt.plot(t, emf(t, a=1 , v=v), label = "v = {}".format(v))
    v*=10
    
plt.title("Emf vs Time: Velocity")
plt.xlabel("Time")
plt.ylabel("EMF")
plt.legend()
```

```python
t = np.linspace(0,10,10000)

def vel(t,q=1,B=1,m=1,g=1):
    return m*g/(q*B)*(np.exp(-q*B*t)-1)

plt.plot(t,vel(t))
plt.title("Velocity vs Time")
plt.xlabel("Time")
plt.ylabel("Velocity")
```
