```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Point Charge ##

n = 5
X, Y = np.mgrid[-n:n, -n:n]

U = X/(X**2 + Y**2 + 0.000001)**(3/2) # added arbitrarily small number so I didnt divide by 0
V = Y/(X**2 + Y**2 + 0.000001)**(3/2) 

plt.quiver(X,Y, U, V, alpha=.5)

plt.xlim(-n, n)
plt.xticks(())
plt.ylim(-n, n)
plt.yticks(())
```

```python
## Dipole ##

n = 5
X, Y = np.mgrid[-n:n, -n:n]

U = X/(X**2 + Y**2 + 0.000001)**(3/2) - X/(X**2 + (Y-2)**2 + 0.000001)**(3/2)
V = Y/(X**2 + Y**2 + 0.000001)**(3/2) - (Y-2)/(X**2 + (Y-2)**2 + 0.000001)**(3/2)

plt.quiver(X, Y, U, V, alpha=.5)

plt.xlim(-n, n)
plt.xticks(())
plt.ylim(-n, n)
plt.yticks(())

```

```python
## Quadrapole AKA 4pole ##

n = 10
X, Y = np.mgrid[-n:n, -n:n]

U = (X+2)/((X+2)**2 + (Y+2)**2 + 0.000001)**(3/2) - (X+2)/((X+2)**2 + (Y-2)**2 + 0.000001)**(3/2) + (X-2)/((X-2)**2 + (Y-2)**2 + 0.000001)**(3/2) - (X-2)/((X-2)**2 + (Y+2)**2 + 0.000001)**(3/2)

V = (Y+2)/((X+2)**2 + (Y+2)**2 + 0.000001)**(3/2) - (Y-2)/((X+2)**2 + (Y-2)**2 + 0.000001)**(3/2) + (Y-2)/((X-2)**2 + (Y-2)**2 + 0.000001)**(3/2) - (Y+2)/((X-2)**2 + (Y+2)**2 + 0.000001)**(3/2)


plt.quiver(X, Y, U, V, alpha=.5)

plt.xlim(-n, n)
plt.xticks(())
plt.ylim(-n, n)
plt.yticks(())


```
