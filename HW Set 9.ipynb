### Problem 3


```python
## Part 3 ##
tetoT = np.linspace(0,np.pi, 100)
tetoI = np.linspace(0,np.pi, 100)
a = 2*np.sqrt(1 - (np.sin(tetoI)/2)**2)/np.cos(tetoI) #n2/n1*(u1/u2), n2/n1 = 2 and u1=u2=1

transmissionimpossible = (2*a/(a+1))
reflexonthehaters = ((1-a)/(a+1))

plt.plot(tetoI, transmissionimpossible, label = "Transmission")
plt.plot(tetoI, reflexonthehaters, label = "Reflection")
plt.plot(tetoI, transmissionimpossible + reflexonthehaters)
#plt.plot(tetoI,a,'.')
plt.title("Recreation of Figure 9.16 From Griffiths")
plt.xlabel("Angle")
plt.legend()
```
